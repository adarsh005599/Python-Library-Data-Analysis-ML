```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# Heat map , cluster map , pivot table heat map

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tipscorr =  tips[['total_bill', 'tip', 'size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot=True)
```

```python
flightcorr = flights[['year', 'passengers']]

```

```python
sns.heatmap(flightcorr.corr(), annot=True, robust=True)
```

```python
sns.clustermap(flightcorr.corr())
```

```python
pivotflights = flights.pivot_table(values='passengers' , index='month', columns='year' )
pivotflights
```

```python
sns.heatmap(pivotflights)
```

```python
sns.lmplot(x= 'data', y='passengers', data = flights)
```
