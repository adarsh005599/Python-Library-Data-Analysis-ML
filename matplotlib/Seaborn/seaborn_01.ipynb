```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# we are now discussing about the distribution plot's like displots/histplot, jointplot, pairplot, rugplot, etc

```python
df = sns.load_dataset('tips')
df['size'].unique()
print(df)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], kde=True , color='gold')

plt.subplot(1,2,2)
sns.histplot(df['tip'], kde=True, color='green')

sns.jointplot(x='total_bill', y='tip', data=df, kind='hist', color='red')
```

```python
sns.pairplot(df, hue='smoker')
```
