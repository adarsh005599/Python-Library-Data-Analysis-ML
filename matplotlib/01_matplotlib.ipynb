### Matplotlib startedü•∂ü•∂

```python
import numpy as np
import matplotlib.pyplot as plt

Swiggy_Booking = np.array(
    [
        [1,1000,2000,3000,4000], 
        [2,1002,2002,3002,4002], 
        [3,1003,2003,3003,4003], 
        [4,1004,2004,3004,4004], 
        [5,1005,2005,3005,4005],
    ]
)

days = Swiggy_Booking[:,0]

plt.figure(figsize=(10,6))
plt.plot(days, Swiggy_Booking[:,1], marker='o', label='Morning Orders')
plt.plot(days, Swiggy_Booking[:,2], marker='o', label='Noon Orders')
plt.plot(days, Swiggy_Booking[:,3], marker='o', label='Evening Orders')
plt.plot(days, Swiggy_Booking[:,4], marker='o', label='Night Orders')

plt.title("Swiggy Orders Per Day")
plt.xlabel("Day")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.legend()
plt.show()
```

## Subplots üê£üê£

```python
plt.subplot(2,2,1)
plt.plot(days, Swiggy_Booking[:,1], marker='o', label='Morning Orders')
```

```python
x= np.array([1,2,3,4,5])
y = x**2

print('===plotting===')

plt.plot(x , y)
plt.title('MJO')
plt.xlabel('number')
plt.ylabel('Mul numbers')
```
