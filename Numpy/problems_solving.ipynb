```python
# Numpy problem solving fileü§ØüòÅ

import  numpy as np
import matplotlib.pyplot as plt

Swiggy_Booking = np.array(
    [
    [1,1000,2000,3000,4000], #1st
    [2,1002,2002,3002,4002], #2nd
    [3,1003,2003,3003,4003], #3rd
    [4,1004,2004,3004,4004], #4th
    [5,1005,2005,3005,4005], #5th
    ]
)

print("===Swiggy Active orders===")
# print(Swiggy_Booking.shape)
# print(Swiggy_Booking.size)
# print("Order details of the first 3: \n", Swiggy_Booking[:3])
# print("Order details of the first 3: \n", Swiggy_Booking[1:,2:4])
# min_seles = np.min(Swiggy_Booking[ : , 1 :] , axis=1)
# print(min_seles)
# print(np.sum(Swiggy_Booking, axis=0))
# Year_booking  = np.sum(Swiggy_Booking[:, 1:], axis=0)
# print(Year_booking)


cumSels = np.cumsum(Swiggy_Booking[:, 1 :], axis=1)
print("Cumulative seles are : \n",cumSels)
# ploting 
print()

print("===plotting the cumulative graph===")

plt.figure(figsize=(8,6))
plt.plot(np.mean(cumSels, axis=0))
plt.title("Average cumulative seles across respective restaurent's :")
plt.xlabel('years')
plt.ylabel("seles/year")
plt.grid(True)
plt.show()
```
